```python
# Importando as bibliotecas necessrias
import matplotlib.pyplot as plt
import pandas as pd

# Dados fictcios de vendas de carros por ms, de janeiro a dezembro
dados_vendas = {
    'Ms': ['Janeiro', 'Fevereiro', 'Maro', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'Vendas': [150, 180, 200, 210, 250, 270, 300, 320, 290, 310, 330, 350]
}

# Criando o DataFrame com os dados
df_vendas = pd.DataFrame(dados_vendas)

# Exibindo os dados
df_vendas.head()

# Gerando o grfico de vendas de carros (no utilizando plt.show())
fig = plt.figure(figsize=(10, 6))
plt.plot(df_vendas['Ms'], df_vendas['Vendas'], marker='o', color='g', label='Vendas de Carros')
plt.title('Vendas de Carros - Concessionria Fictcia (2025)', fontsize=14)
plt.xlabel('Ms', fontsize=12)
plt.ylabel('Quantidade de Vendas', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Retornando a figura
fig
```
