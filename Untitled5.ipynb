```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt
import pandas as pd

# Dados fictícios de vendas de caminhonetes por mês, de janeiro a dezembro (com mais variação)
dados_vendas_caminhonetes = {
    'Mês': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'Vendas': [80, 70, 95, 130, 180, 220, 250, 300, 270, 280, 320, 350]  # Variação maior nas vendas
}

# Criando o DataFrame com os dados
df_vendas_caminhonetes = pd.DataFrame(dados_vendas_caminhonetes)

# Exibindo os dados
df_vendas_caminhonetes.head()

# Gerando o gráfico de vendas de caminhonetes (não utilizando plt.show())
fig = plt.figure(figsize=(10, 6))
plt.plot(df_vendas_caminhonetes['Mês'], df_vendas_caminhonetes['Vendas'], marker='o', color='r', label='Vendas de Caminhonetes')
plt.title('Vendas de Caminhonetes - Concessionária Fictícia (2025)', fontsize=14)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Quantidade de Vendas', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Retornando a figura
fig
```
