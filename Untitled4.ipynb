```python
# Importando as bibliotecas necessrias
import matplotlib.pyplot as plt
import pandas as pd

# Dados fictcios de vendas de motos por ms, de janeiro a dezembro (com mais variao)
dados_vendas_motos = {
    'Ms': ['Janeiro', 'Fevereiro', 'Maro', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'Vendas': [120, 110, 90, 150, 200, 180, 210, 250, 230, 190, 270, 300]
}

# Criando o DataFrame com os dados
df_vendas_motos = pd.DataFrame(dados_vendas_motos)

# Exibindo os dados
df_vendas_motos.head()

# Gerando o grfico de vendas de motos (no utilizando plt.show())
fig = plt.figure(figsize=(10, 6))
plt.plot(df_vendas_motos['Ms'], df_vendas_motos['Vendas'], marker='o', color='b', label='Vendas de Motos')
plt.title('Vendas de Motos - Concessionria Fictcia (2025)', fontsize=14)
plt.xlabel('Ms', fontsize=12)
plt.ylabel('Quantidade de Vendas', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Retornando a figura
fig
```
